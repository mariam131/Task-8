# **Line Plot For Age Distribution**

```python
age_groups =df_arabs['Q1'].value_counts().sort_index()
age_groups
```

```python
x= age_groups.index
y= age_groups.values
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(10,7))
plt.plot(x,y)
plt.title('Age Distribution')
plt.xlabel('Age groups')
plt.ylabel('frequency')
plt.show()
```

# **Arab kaggle users by programming language Distribution**

```python
df_arabs
```

```python
df.columns # احنا عايزين ناخد كل الصفوف اللي فيها رقم 7
```

```python
q7_coloumns= df.columns[df.columns.str.contains('^Q7')]  # ^  هذه العلامة بمعني ان تكون في بداية الكلمة
q7_coloumns
```

```python
df_arabs[q7_coloumns]
```

```python
df['Q7_Part_1'].value_counts().index
```

```python
dict_7 = dict()
for col in q7_coloumns:
    key = df_arabs[col].value_counts().index[0]
    dict_7[key]= df_arabs[col].value_counts()[0]
    
dict_7
```

```python
q7_series = pd.Series(dict_7)
q7_series
```

```python
plt.figure(figsize=(10,10))
plt.pie(q7_series , labels = q7_series.index , autopct= '%1.1f%%')
plt.legend(q7_series.index)
plt.show()
```

```python
fig,axes= plt.subplots(1,2, figsize = (20,8))
fig.patch.set_facecolor('#F2F2F2')

axes[0].bar(q7_series.index , q7_series.values)
axes[0].set_xlabel('programming language')
axes[0].set_ylabel('count')
axes[0].tick_params(labelrotation=80 , labelsize= 15)
axes[0].spines['left'].set_visible('green')
axes[0].spines['bottom'].set_visible('black')


axes[1].pie(q7_series , labels = q7_series.index , autopct= '%1.1f%%')

plt.title('programming language distribution of kaggle users')
plt.show()
```
