```python
# lession 1
# Line plot (sns.relplot)
```

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
flights = sns.load_dataset("flights")
flights
```

```python
flights_feb = flights[flights['month']=='Feb']
flights_feb
```

```python
sns.set_style("darkgrid")
sns.lineplot( x ='year', y ='passengers' ,data =flights_feb)
plt.title("passengers vs year")
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.lineplot( x ='year', y ='passengers' ,data =flights , hue ='month')
```

```python
fmri = sns.load_dataset("fmri")
fmri
```

```python
sns.lineplot( x ='timepoint', y ='signal' ,data =fmri , hue='event', style='region')
```

```python
sns.relplot( x ='timepoint', y ='signal' ,data =fmri , hue='event', style='region', kind='line', col='region',row='event')
```

```python
# lesson 2
# scatter plot   وظيفته رسم العلاقة بين متغيرين رقمين
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('darkgrid')
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
sns.scatterplot(data=tips , x='total_bill' , y='tip')
```

```python
sns.scatterplot(data=tips , x='total_bill' , y='tip', hue='sex' , style= 'smoker', size='size')
```

```python
sns.scatterplot(data=tips , x='total_bill' , y='tip', hue='time',style="time",  markers={"Lunch":"s", "Dinner":"D"})
```

```python
sns.relplot(data=tips , x='total_bill' , y='tip', hue='time', style='sex', kind='scatter', col='sex' , row='day')
```

```python
# lesson 3
# Regreesion plots (sns.implot)   (categorical)يمكن رسم بين قيمتين رقمين او قيمة رقمية و
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('dark')
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
sns.regplot(data=tips , x='total_bill' , y='tip')
```

```python
sns.regplot(data=tips , x='size' , y='tip')
```

```python
sns.regplot(data=tips , x='size' , y='tip' , x_jitter=0.1) # x_jutter للفصل بين القيم قليلا لفهمها
```

```python
sns.regplot(data=tips , x='size' , y='tip' , x_estimator=np.median) #  لحساب المتوسط
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue="smoker", col="smoker", row="sex", truncate=False)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

penguins = sns.load_dataset("penguins")
penguins.head()
```

```python
# sns.histplot
```

```python
sns.histplot(penguins , x='flipper_length_mm')
```

```python
sns.histplot(penguins , x='flipper_length_mm' , bins =20)
plt.show()
```

```python
sns.histplot(penguins , x='flipper_length_mm' , stat ='density')
```

```python
sns.histplot(penguins, x="species", stat="density")
```

```python
sns.histplot(penguins, x="species", stat="density" , shrink=0.8) # shrink  تفصل بين كل عمود والاخر بمقدار 0.8
```

```python
sns.histplot(penguins , x='flipper_length_mm', hue ='species')
```

```python
sns.histplot(penguins , x='flipper_length_mm', hue ='species', element='step') # element =step  لاخفاء الخطوط بين الاعمدة
```

```python
sns.histplot(penguins , x='flipper_length_mm' , stat ='density' , kde =True) # kde = true لرسم المنحني فوق الشكل
```

```python
sns.histplot(penguins , x='flipper_length_mm' , hue = 'species' , kde =True)
```

```python
sns.histplot(data= penguins , y='flipper_length_mm')
```
