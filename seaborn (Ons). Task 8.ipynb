```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
flights = sns.load_dataset("flights")
flights
```

```python
flights_feb = flights[flights['month']=='Feb']
flights_feb
```

```python
sns.set_style("darkgrid")
sns.lineplot( x ='year', y ='passengers' ,data =flights_feb)
plt.title("passengers vs year")
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.lineplot( x ='year', y ='passengers' ,data =flights , hue ='month')
```

```python
fmri = sns.load_dataset("fmri")
fmri
```

```python
sns.lineplot( x ='timepoint', y ='signal' ,data =fmri , hue='event', style='region')
```

```python
sns.relplot( x ='timepoint', y ='signal' ,data =fmri , hue='event', style='region', kind='line', col='region',row='event')
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('darkgrid')
```

```python
sns.scatterplot(data=tips , x='total_bill' , y='tip', hue='sex' , style= 'smoker', size='size')
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('dark')
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
sns.regplot(data=tips , x='total_bill' , y='tip')
```

```python
sns.regplot(data=tips , x='size' , y='tip')
```

```python
sns.regplot(data=tips , x='size' , y='tip' , x_jitter=0.1) # x_jutter Ù„Ù„ÙØµÙ„ Ø¨ÙŠÙ† Ø§Ù„Ù‚ÙŠÙ… Ù‚Ù„ÙŠÙ„Ø§ Ù„ÙÙ‡Ù…Ù‡Ø§
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue="smoker", col="smoker", row="sex", truncate=False)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

penguins = sns.load_dataset("penguins")
penguins.head()
```
